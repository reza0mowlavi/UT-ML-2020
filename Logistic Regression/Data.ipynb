```python
from logistic_regression import StratifiedShuffleSplit

X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X = np.array(X)
y = np.where(y==' <=50K',0,1).astype('int64')

split = StratifiedShuffleSplit(test_size=0.3,shuffle=True)
(X_train,y_train),(X_test,y_test) = split.split(X,y)
```

### Create baseline model
#### It assign all samples to negative "<50" class.

```python
class Baseline:
    def fit(self,X,y):
        return self
    
    def predict(self,X):
        return np.zeros(X,dtype='int64')
    
    def score(self,X,y):
        pred = self.predict(np.shape(X)[0])
        return self.accuracy(y, pred)

    def accuracy(self, y_true, y_pred):
        T = np.sum(y_true == y_pred)
        return T / np.shape(y_true)[0]
    
baseline_model = Baseline()
baseline_model.fit(X_train,y_train)
print(f'Train Acc:{baseline_model.score(X_train,y_train)} -- Test Acc:{baseline_model.score(X_test,y_test)}')
```
